```python
# '일시' 열을 datetime 형식으로 변환
train['일시'] = pd.to_datetime(train['일시'])

# 연도, 월, 일 열 생성
train['연도'] = train['일시'].dt.year
train['월'] = train['일시'].dt.month
train['일'] = train['일시'].dt.day

# '일시' 열을 datetime 형식으로 변환
submission['일시'] = pd.to_datetime(submission['일시'])

# 연도, 월, 일 열 생성
submission['연도'] = submission['일시'].dt.year
submission['월'] = submission['일시'].dt.month
submission['일'] = submission['일시'].dt.day
```

```python
import pandas as pd

# '일시' 열을 datetime 형식으로 변환
train['일시'] = pd.to_datetime(train['일시'])

# '일시' 열에서 주차 정보 추출
train['주차'] = train['일시'].dt.isocalendar().week

# 결과 확인
print(train[['일시', '주차']])

# '일시' 열을 datetime 형식으로 변환
submission['일시'] = pd.to_datetime(submission['일시'])

# '일시' 열에서 주차 정보 추출
submission['주차'] = submission['일시'].dt.isocalendar().week

# 결과 확인
print(submission[['일시', '주차']])
```

```python
# 결측치 확인
missing_values = train.isnull()

# 각 열별로 결측치 개수 세기
missing_count_per_column = missing_values.sum()

# 결과 출력
print(missing_count_per_column)
```

```python
# '일시' 열을 datetime 형식으로 변환
train['일시'] = pd.to_datetime(train['일시'])

# '일시' 열에서 날짜만 추출하여 '년도-월-일' 형식으로 저장
train['날짜'] = train['일시'].dt.strftime('%Y-%m-%d')

# '날짜' 열을 datetime 형식으로 변환
train['날짜'] = pd.to_datetime(train['날짜'])

# '날짜' 열에서 1년 중 몇 번째 날인지 추출
train['년중_일수'] = (train['날짜'] - pd.to_datetime(train['날짜'].dt.year, format='%Y')).dt.days + 1

# 결과 확인
print(train[['일시', '주차', '년중_일수']])



# '일시' 열을 datetime 형식으로 변환
submission['일시'] = pd.to_datetime(submission['일시'])

# '일시' 열에서 날짜만 추출하여 '년도-월-일' 형식으로 저장
submission['날짜'] = submission['일시'].dt.strftime('%Y-%m-%d')

# '날짜' 열을 datetime 형식으로 변환
submission['날짜'] = pd.to_datetime(submission['날짜'])

# '날짜' 열에서 1년 중 몇 번째 날인지 추출
submission['년중_일수'] = (submission['날짜'] - pd.to_datetime(submission['날짜'].dt.year, format='%Y')).dt.days + 1

# 결과 확인
print(submission[['일시', '주차', '년중_일수']])
```

```python
# 계절 변수 생성
train['계절'] = ''

for i, month in enumerate(train['월']):
    if 3 <= month <= 5:
        train.at[i, '계절'] = '봄'
    elif 6 <= month <= 8:
        train.at[i, '계절'] = '여름'
    elif 9 <= month <= 11:
        train.at[i, '계절'] = '가을'
    else:
        train.at[i, '계절'] = '겨울'

# 결과 확인
print(train[['일시', '월', '계절']])


# 계절 변수 생성
submission['계절'] = ''

for i, month in enumerate(submission['월']):
    if 3 <= month <= 5:
        submission.at[i, '계절'] = '봄'
    elif 6 <= month <= 8:
        submission.at[i, '계절'] = '여름'
    elif 9 <= month <= 11:
        submission.at[i, '계절'] = '가을'
    else:
        submission.at[i, '계절'] = '겨울'

# 결과 확인
print(submission[['일시', '월', '계절']])
```

```python
# '계절' 열에 대한 원핫인코딩 수행
one_hot_encoded = pd.get_dummies(train['계절'], prefix='계절')

# 기존 데이터프레임과 합치기
train = pd.concat([train, one_hot_encoded], axis=1)

# '계절' 열과 중복된 열 삭제
train.drop(['계절'], axis=1, inplace=True)

# 결과 확인
print(train)


# '계절' 열에 대한 원핫인코딩 수행
one_hot_encoded = pd.get_dummies(submission['계절'], prefix='계절')

# 기존 데이터프레임과 합치기
submission = pd.concat([submission, one_hot_encoded], axis=1)

# '계절' 열과 중복된 열 삭제
submission.drop(['계절'], axis=1, inplace=True)

# 결과 확인
print(train)
print(submission)
```

```python
import optuna
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor

# 필요한 변수 선택 (예시로 선택된 변수들이므로 실제로 사용하는 변수에 맞게 수정 필요)
selected_features = ['연도', '월', '일', '주차', '년중_일수', '계절_가을', '계절_겨울', '계절_봄', '계절_여름']
target_variable = '평균기온'

# 선택된 변수들을 사용하여 훈련 데이터셋 준비
X_train = train[selected_features]
y_train = train[target_variable]



# 모델 정의
rf = RandomForestRegressor(random_state=2024,
                            n_estimators=1000,
                            max_depth=10,
                            min_samples_leaf=3,
                            verbose=100)  # verbose 값을 원하는 값으로 설정

# 모델 학습
rf.fit(X_train, y_train)
```

```python
# submission 데이터에서 필요한 변수만 남기기 (평균기온을 예측하기 위한 변수)
submission_data = submission[['연도', '월', '일', '주차', '년중_일수', '계절_가을', '계절_겨울', '계절_봄', '계절_여름']]

# 예측 수행
submission_pred = rf.predict(submission_data)
```

```python
submission = pd.read_csv('sample_submission.csv')
submission['평균기온']=submission_pred
```
