```python
# 결측치 확인
missing_values = train.isnull()

# 각 열별로 결측치 개수 세기
missing_count_per_column = missing_values.sum()

# 결과 출력
print(missing_count_per_column)
```

```python
# '일시' 열을 datetime 형식으로 변환
train['일시'] = pd.to_datetime(train['일시'])

# '일시' 열에서 날짜만 추출하여 '년도-월-일' 형식으로 저장
train['날짜'] = train['일시'].dt.strftime('%Y-%m-%d')

# '날짜' 열을 datetime 형식으로 변환
train['날짜'] = pd.to_datetime(train['날짜'])

# '날짜' 열에서 1년 중 몇 번째 날인지 추출
train['년중_일수'] = (train['날짜'] - pd.to_datetime(train['날짜'].dt.year, format='%Y')).dt.days + 1

# 결과 확인
print(train[['일시', '주차', '년중_일수']])



# '일시' 열을 datetime 형식으로 변환
submission['일시'] = pd.to_datetime(submission['일시'])

# '일시' 열에서 날짜만 추출하여 '년도-월-일' 형식으로 저장
submission['날짜'] = submission['일시'].dt.strftime('%Y-%m-%d')

# '날짜' 열을 datetime 형식으로 변환
submission['날짜'] = pd.to_datetime(submission['날짜'])

# '날짜' 열에서 1년 중 몇 번째 날인지 추출
submission['년중_일수'] = (submission['날짜'] - pd.to_datetime(submission['날짜'].dt.year, format='%Y')).dt.days + 1

# 결과 확인
print(submission[['일시', '주차', '년중_일수']])
```

```python
# 계절 변수 생성
train['계절'] = ''

for i, month in enumerate(train['월']):
    if 3 <= month <= 5:
        train.at[i, '계절'] = '봄'
    elif 6 <= month <= 8:
        train.at[i, '계절'] = '여름'
    elif 9 <= month <= 11:
        train.at[i, '계절'] = '가을'
    else:
        train.at[i, '계절'] = '겨울'

# 결과 확인
print(train[['일시', '월', '계절']])


# 계절 변수 생성
submission['계절'] = ''

for i, month in enumerate(submission['월']):
    if 3 <= month <= 5:
        submission.at[i, '계절'] = '봄'
    elif 6 <= month <= 8:
        submission.at[i, '계절'] = '여름'
    elif 9 <= month <= 11:
        submission.at[i, '계절'] = '가을'
    else:
        submission.at[i, '계절'] = '겨울'

# 결과 확인
print(submission[['일시', '월', '계절']])
```

```python
# '계절' 열에 대한 원핫인코딩 수행
one_hot_encoded = pd.get_dummies(train['계절'], prefix='계절')

# 기존 데이터프레임과 합치기
train = pd.concat([train, one_hot_encoded], axis=1)

# '계절' 열과 중복된 열 삭제
train.drop(['계절'], axis=1, inplace=True)

# 결과 확인
print(train)


# '계절' 열에 대한 원핫인코딩 수행
one_hot_encoded = pd.get_dummies(submission['계절'], prefix='계절')

# 기존 데이터프레임과 합치기
submission = pd.concat([submission, one_hot_encoded], axis=1)

# '계절' 열과 중복된 열 삭제
submission.drop(['계절'], axis=1, inplace=True)

# 결과 확인
print(train)
print(submission)
```

```python
from sklearn.model_selection import GridSearchCV
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# 데이터를 훈련셋과 검증셋으로 나누기
X_train, X_val, y_train, y_val = train_test_split(selected_data.drop('평균기온', axis=1), selected_data['평균기온'], test_size=0.1, random_state=42)

# 병렬 처리 활성화
model = CatBoostRegressor(random_seed=42, od_type='Iter', od_wait=20, grow_policy='Lossguide', bootstrap_type='MVS')

# CatBoostRegressor의 가능한 하이퍼파라미터 범위 설정
param_grid = {
    'iterations': [500, 1000, 1500],
    'depth': [4, 6, 8, 10],
    'learning_rate': [0.001, 0.01, 0.1],
    'bagging_temperature': [0.1, 1.0, 10.0],
    'l2_leaf_reg': [1e-4, 1e-2, 1e2],
    'subsample': [0.1, 0.5, 1.0],
    'colsample_bylevel': [0.1, 0.5, 1.0],
    'min_child_samples': [1, 10, 20],
    'max_leaves': [10, 30, 50],
}

# RandomizedSearchCV를 사용하여 최적의 하이퍼파라미터 찾기
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, scoring='neg_mean_absolute_error', cv=3, n_iter=10, verbose=1, n_jobs=-1)
random_search.fit(X_train, y_train)

# 최적의 하이퍼파라미터 확인
best_params = random_search.best_params_
print("Best Parameters:", best_params)

# 최적의 하이퍼파라미터로 모델 재훈련
best_model = CatBoostRegressor(
    random_seed=42,
    od_type='Iter',
    od_wait=20,
    grow_policy='Lossguide',
    **best_params
)

best_model.fit(X_train, y_train, eval_set=[(X_val, y_val)], early_stopping_rounds=20, verbose=True)
```

```python
# submission 데이터에서 필요한 변수만 남기기 (평균기온을 예측하기 위한 변수)
submission_data = submission[['연도', '월', '일', '주차', '년중_일수', '계절_가을', '계절_겨울', '계절_봄', '계절_여름']]

# 예측 수행
submission_pred = best_model.predict(submission_data)
```
